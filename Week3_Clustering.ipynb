```python
from bs4 import BeautifulSoup
import xml
results = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(results, 'lxml')
```

```python
table = soup.find('table')
fields = table.find_all('td')

postcode = []
borough = []
neighborhood = []

for i in range(0, len(fields), 3):
    postcode.append(fields[i].text.strip())
    borough.append(fields[i+1].text.strip())
    neighborhood.append(fields[i+2].text.strip())

df = pd.DataFrame([postcode, borough, neighborhood]).transpose()
df.columns = ['Postcode', 'Borough', 'Neighborhood']
df.head()
```

```python
df['Borough'].replace('Not assigned', np.nan, inplace = True)
df.dropna(subset = ['Borough'], inplace = True)
df.head()
```

```python
df_results = df.groupby(['Postcode', 'Borough'])['Neighborhood'].apply(', '.join).reset_index()
df_results.columns = ['Postcode', 'Borough', 'Neighborhood']
df_results
```

```python
df_results['Neighborhood'].replace('Not assigned', 'Queen''s Park', inplace = True)
df_results
```

Part 2 

```python
df_geo = pd.read_csv('http://cocl.us/Geospatial_data')
df_geo.columns = ['Postcode', 'Latitude', 'Longitude']
```

```python
df_all = pd.merge(df_results, df_geo, on = ['Postcode'], how = 'inner')
df_sol = df_all[['Postcode', 'Borough', 'Neighborhood', 'Latitude', 'Longitude']].copy()

df_sol.head(11)
```

Part 3

```python
a = 'Toronto, Canada'

geo = Nominatim()
loc = geo.geocode(a)
lat = loc.latitude
long = loc.longitude

map_t = folium.Map(location = [lat, long], zoom_start = 8)

for lat,  lng, b, n in zip(df_sol['Latitude'], df_sol['Longitude'], df_sol['Borough'], df_sol['Neighborhood']):
    label = '{}, {}'.format(n, b)
    label = folium.Popup(label, parse_html  = True)
    folium.CircleMarker([lat, lng], radius =  3, popup = label, color = 'blue', fill = True, parse_html = False).add_to(map_t)
    
map_t
```
